```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=.1) #how much the coefficients are shrunk to account for colinearity
```

```python
predictors = weather.columns[~weather.columns.isin(["target","name","station"])]
predictors
```

```python
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []  # each element in the list is a dataframe that had predictions for 90 days
    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i, :]  # all data before current row
        test = weather.iloc[i:(i + step), :]
        model.fit(train[predictors], train["target"])
        pred = model.predict(test[predictors])
        pred = pd.Series(pred, index=test.index)  # series is easier to work with than numpy array
        combine = pd.concat([test["target"], pred], axis=1)

        combine.columns = ["actual", "prediction"]
        combine["difference"] = (combine["prediction"] - combine["actual"]).abs()
        all_predictions.append(combine)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(weather, rr, predictors)
predictions
```

```python
print(backtest(weather, rr, predictors))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(predictions["actual"], predictions["prediction"])
```

```python
predictions["difference"].mean()
#on average we are 5 degrees off
```

```python
#improve and make accuracy better
#calculating average temperature and precipitation of past 3 days
def pct_diff(old,new):
  return(new - old)/old #to calculate % difference

def compute_rolling(weather , horizon,col): #find rolling averages
  weather = pd.DataFrame(weather.columns)
  #weather = pd.DataFrame(columns=["tmax", "tmin", "prcp"])
  label=f"rolling_{horizon}_{col}"# format string; used to create dynamic strings
  weather[label] = weather[col].rolling(horizon).mean()
  weather[f"{label}_pct"] = pct_diff(weather[label],weather[col])
  return weather

rolling_horizon = [3, 14]

for horizon in rolling_horizon:
    for col in ["tmax", "tmin", "prcp"]:
        weather = compute_rolling(weather, horizon, col)

```

```python
#adding more predictors
# Assuming your index is not a DatetimeIndex
weather.index = pd.to_datetime(weather.index)

def expand_mean(df):
    return df.expanding(1).mean()

for col in ["tmax", "tmin", "prcp"]:
    weather[f"month_avg_{col}"] = weather[col].groupby(weather.index.month, group_keys=False).apply(expand_mean)
    weather[f"day_avg_{col}"] = weather[col].groupby(weather.index.day_of_year, group_keys=False).apply(expand_mean)
```
