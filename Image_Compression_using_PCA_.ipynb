```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
img =  cv2.cvtColor(cv2.imread("image3.jpg"), cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
r,g,b=cv2.split(img)
```

```python
#0 is the minimum value a colour can have
#255 is the maximum value a colour can have
#we do this inorder to get values between 0 and 1
r,g,b=r/255, g/255, b/255
```

```python
plt.imshow(r)
```

```python
plt.imshow(g)
```

```python
plt.imshow(b)
```

```python
pca_components=300
pca_r= PCA(n_components=pca_components)
reduced_r= pca_r.fit_transform(r)
```

```python
pca_g= PCA(n_components=pca_components)
reduced_g= pca_g.fit_transform(g)
```

```python
pca_b= PCA(n_components=pca_components)
reduced_b= pca_b.fit_transform(b)
```

```python
reduced_r.shape
reduced_g.shape
reduced_b.shape
```

```python
combined = np.array([reduced_r,reduced_g,reduced_b])
```

```python
reconstructed_r=pca_r.inverse_transform(reduced_r)
reconstructed_g=pca_g.inverse_transform(reduced_g)
reconstructed_b=pca_b.inverse_transform(reduced_b)
```

```python
reconstructed_r.shape
reconstructed_g.shape
reconstructed_b.shape
```

```python
plt.imshow(reconstructed_r)
```

```python
plt.imshow(reconstructed_g)
```

```python
plt.imshow(reconstructed_b)
```

```python
img_reconstructed = (cv2.merge((reconstructed_r,reconstructed_g,reconstructed_b)))
```

```python
plt.imshow(img_reconstructed)
```
